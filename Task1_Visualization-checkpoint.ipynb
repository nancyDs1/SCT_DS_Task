```python
# Load a libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Data set is from "https://censusindia.gov.in/census.website/data/census-tables"
# Data cleaning have done using Excel
```

```python
# load a dataset

df = pd.read_csv("surat_population.csv")
df
```

```python
# cut the total column
df.drop(columns = ["total"])
```

```python
# cut the total row
df = df.drop([7],axis = 0)
df
```

```python
import matplotlib.ticker as mticker


# Seaborn Enhanced Line Chart for Population Variation by Age Group
plt.figure(figsize=(10, 4))
sns.lineplot(x=df["age"], y=df["male"]/1000000, color="green", linewidth=2.5, label="Male")
sns.lineplot(x=df["age"], y=df["female"]/1000000, color="m", linewidth=2.5, label="Female")
plt.fill_between(df["age"], df["male"]/1000000, alpha=0.2, color="green")
plt.fill_between(df["age"], df["female"]/1000000, alpha=0.2, color="m")

# Format Y-axis labels as "0.1M", "0.5M", etc.
def millions_formatter(x, pos):
    return f"{x:.1f}M"

plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(millions_formatter))

plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Population Count", fontsize=12)
plt.title("Population Variation by Age Group (in millions)", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=12, title="Gender")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
